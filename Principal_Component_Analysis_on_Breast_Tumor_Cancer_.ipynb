# Principal Component Analysis

* This is an unsupervised machine learning algorithm that is used mostly for dimensionality reduction, however this analysis will achieve that and provide biologically relevant insight
* It is basically a transformation of the data that attempts to find out what features explains the most variance in the data
* ...i.e. get rid of the components that minimally explain variance in the data
* Example, say from 50 features, we reduce to 10 (features/Principal Components), the first principal components explains the most variance, the second explains the most remaining variance, and so on
* ... this transformation does not reduce the number of rows
* These 10 PCs are not in 1-to-1 correspondence to the original features, but rather a linear combination of the original features
* For example, if we have original features x₁, x₂, and x₃, the first principal component might be:
    * PC₁ = 0.6x₁ + 0.3x₂ + 0.1x₃
    * The coefficients (0.6, 0.3, 0.1) form the eigenvector, and they show how much each original feature contributes to that component
    
    
* **Pros:**
    * PCA is commonly used for data visualization -where high dimensional data is reduced to 2D or 3D for plotting
    * Valuable for noise reduction (components explaining minimal variance of data are assumed to be noise)
    * PCA simplifies complex dataset and speed up machine learning algorithms while retaining most important information


* **Cons/Limitations:**
    * PCA assumes linear relationship between variables
    * PCA assumes variance == importance (hmmm... but is this always the case? )
    * Difficult to interpret components
    * Sensitive to outliers, hence features must be scaled
    
    
* <span style="color: red;">NOTE</span>: PCA works best when original features are correlated. If the average correlation is 0.4 or more, then PCA will be beneficial


* **How can we determine the optimal number of Principal Components for PCA?**
   * The overaching goal here is to look at how much total variance is explained as we add more components
   * There are several methods to do this including the elbow method.
   * Domain knowledge is needed to determine the best method
   * For the sake of simplicity and brivity, I will use the popular method: <span style="color: red;"> (Cumulative) Explained Variance Ratio</span>

# Breast Cancer Dataset
* This dataset has <span style="color: red;">30 features</span> and I have explained it in a previous project in my [GitHub repo](https://github.com/Benjamin-compute/Breast-Cancer-Prediction__Comparing-4-ML-Algorithms)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score


%matplotlib inline
```

```python
# get data

df = pd.read_csv("breast_cancer_data")
```

```python
df.head()
```

```python
# drop irrevalant columns
# * id
# * diagnosis (this is the target variable)
# * Unnamed:32

df_features = df.drop(['id', 'diagnosis', 'Unnamed: 32'], axis= 1)

df_features.info()
```

```python
# See the correlation of the dataset

plt.figure(figsize = (10,8))

corr_matrix = df_features.corr()

sns.heatmap(corr_matrix, center = 0)
```

```python
# For ease, check if the average absolute correlation is above 0.4

# Get average correlation (excluding diagonal)

avg_corr = np.abs(corr_matrix.values[np.triu_indices_from(corr_matrix.values, 1)]).mean()


print(f"Average correlation: {avg_corr:.2f}")
```

### What does this means?
* A lot of these tumor features move together
* Such as: an increase in radius, would lead to increase in several other variables like perimeter, area, concavity etc



### Insight:
* We do not need all 30 measurements to understand tumor structure because some are redundant
* Since the average absolute correlation is ~0.4, we assume PCA will be beneficial to use
* With PCA, we can develop a cheaper diagnostic panels without lossing much biological signal
* Thus, we proceed

```python
# currently we are dealing with a 30 dimensional data (which is not visually feasible).
# Note: 30 dimensional == 30 features/columns in df

# this is where PCA comes in.
```

```python
# We will scale the data first,
# so that each feature has a single unit variance
# before we actually use PCA on it


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
# fit data

scaler.fit(df_features)
```

```python
# apply transformation

scaled_features = scaler.transform(df_features)
```

```python
# now perform PCA

from sklearn.decomposition import PCA


pca = PCA()     # defaultly using all components (30)


# fit all features/components
pca.fit(scaled_features)
```

### Find the Optimal Number of Principal Components

<font color="red">Rule of thumb:</font> Choose *n_components* that explain 80-95% of the cummulative variance.


```python
# Calculate cumulative explained variance
cumulative_variance = np.cumsum(
                                    pca.explained_variance_ratio_
                                )
```

```python
# Plot this "cumulative_variance"

# create figure size
plt.figure(figsize=(10, 6))

# create the markers (blocked circles)
plt.plot(range(1, len(cumulative_variance) + 1),
         cumulative_variance,
         'bo-')

# marking out the variance
plt.axhline(y=0.95, color='r', linestyle='--', label='95% variance')
plt.axhline(y=0.90, color='g', linestyle='--', label='90% variance')
plt.axhline(y=0.80, color='y', linestyle='--', label='80% variance')

# labels
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Components')
plt.legend()
plt.grid(True)
```

# Insights:
* From the plot, PCA will explain $ 95\% $ variance of the dataset using <font color="red">10 components</font>
* With this number of components we can be sure that we have a balance of: feature size and speed of the algorithm (PCA)

```python
# For validation:

n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1

print(f"Number of components needed for a 95% variance = {n_components_95}")
```

```python
# Now we re-run our PCA now using "n_components = 10"


pca_verified = PCA(n_components= 10)

# fit data
pca_verified.fit(scaled_features)

# transform
principal_components = pca_verified.transform(scaled_features)
```

```python
# Create a DataFrame with the principal components

df_principal_components = pd.DataFrame(
                                    data= principal_components,
                                    columns= ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10']
                                    )
```

```python
df_principal_components.head()
```

```python
# some checks

print(f"Shape of dataset with original features: {df_features.shape} ")
print('\n')
print(f"Shape of dataset with PCA transformed features: {df_principal_components.shape} ")
```

# How much variance was explained by each component?

```python
# get the 'explained_variance_ratio'
variance_explained = pca_verified.explained_variance_ratio_

variance_explained
```

```python
# Quick Check: the above variance should sum up to 95% or more

sum(variance_explained)
```

```python
print('\n')

print("Variance explained by each component:")

print('\n')

# iterate through and calculate
for i, var in enumerate(variance_explained):
    print(f"PC{i+1}: {var:.3f} ({var*100:.1f}%)")
```

# Insights on these 10 Components
* These 10 components explain 95% of all variation
* Of which <font color="red">63% is explained by the first TWO components</font>: This means, most of the tumor story is hidden here


####  Do we know EXACTLY what these new principal components represent?
* Unfortunately, principal components do not have meaningful column names like the original features. This is why we just named them like PC1, PC2, PC3, etc.
* However, we can interpret what they represent by examining the loadings (weights):

```python
df_principal_components.columns
```

```python
# Get the loadings (how much each original feature contributes to the PCs)
loadings = pca_verified.components_

# Create a DataFrame to see loadings clearly
loadings_df = pd.DataFrame(
                            loadings.T,    # Transpose to get features as rows
                            columns= df_principal_components.columns,
                            index=   df_features.columns  # Original feature names
                            )

print("Loadings (contribution of each original feature to the PCs):")
print(loadings_df)
```

### Drill Down on the first 2 Components to draw some insights

#### * PC1:

```python
# To interpret PC1, look at which features have the highest loadings
print("PC1 interpretation:")

# map features to patterns
pc1_loadings = loadings_df['PC1'].abs().sort_values(ascending=False)

print(pc1_loadings.head(7))
```

# Insights:

* The top weighted loadings of the PC1 have some things in common:
  * They <u>describe tumor edge irregularity and shape distortion</u>
  * They DO NOT describe size or texture
  * The component could help answer questions around tumor boundary (with possible focus on concavity)
  
* <font color="red">Conclusion: Since PC1 explained 44% of the total variation, we can say, **the dominant signal** in the breast cancer tumor is likely <u>Shape Irregularity</u> </font>

#### * PC2:

```python
# To interpret PC2, look at which features have the highest loadings
print("PC2 interpretation:")

# map features to patterns
pc2_loadings = loadings_df['PC2'].abs().sort_values(ascending=False)

print(pc2_loadings.head(7))
```

## Insights
* PC2 is dominated by:
  * <font color="green">fractal dimension </font> which measures:
    * Boundary complexity across scales
    * How rough the tumor edge is
  * <font color="green">Size </font> which mixes:
    * overall scale
* <font color="red">We can conclude that PC2 is telling us <u>Tumor size, in relation to, structural complexity</u> </font>

<br>

* Possibly, the overaching question PC2 could help us answer: <u>Is the tumor structurally intricate or biologically complex? </u>

# Biological Discovery
* Test PC1 & PC2 against diagnosis (Malignant/Benign)


```python
# Tumor cordinates

pc_df = pd.DataFrame(principal_components[:, :2], columns=['PC1', 'PC2'])    # principal compenents it the matrix transformed for just 10 comp. (See cell 15 of this script)
pc_df['diagnosis'] = df['diagnosis'].values

pc_df.head()
```

```python
# Visualize tumor distribution

colors = pc_df['diagnosis'].map({'M': 'red',
                                 'B': 'blue'})

plt.figure(figsize=(8,6))
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=colors, alpha=0.7)
plt.xlabel('PC1 (Shape/Structural Distortion)')
plt.ylabel('PC2 (Complexity)')
plt.title('PCA Space of Tumor Distribution')
plt.show()
```

#  Insights:

* The more structurally distorted a tumor is (PC1 or x axis), the more likely it is malignant
* Overlaps: We have more vertical overlap than horizontal overlap OR the data is more spread out horizontally
  * This tells internal complexity (vertical overlap, PC2) alone does not determine malignancy, shape/structural distortion (horizontal overlap, PC1) does
* Malignant tumors are more heterogenous morphologically
* <u>Therefore, we can summarize that cancer introduces more structural variability than benign growth</u>

# How do these relates to Breast Tumor Diagnosis?

```python
diag_numeric = pc_df['diagnosis'].map({'M':1, 'B':0})

print("Correlation PC1 vs Diagnosis:", pc_df['PC1'].corr(diag_numeric))
print("Correlation PC2 vs Diagnosis:", pc_df['PC2'].corr(diag_numeric))
```

# Insights
* PC1 vs Diagnosis: correlation ≈ 0.8 (80%)
  * This means that shape irregualarity or distortion drives malignancy (cancerous tumor)
  * Tumors with higher PC1 scores are much more likely to be malignant

* PC2 vs Diagnosis: correlation ≈ -0.2 (-20%)
  * Structural complexity (on its own) does not strongly determine whether a tumor is malignant


# Which Pattern is more predictive?


```python
# Using a logistic regression model
model = LogisticRegression()
model.fit(pc_df[['PC1','PC2']], diag_numeric)

preds = model.predict_proba(pc_df[['PC1','PC2']])[:,1]
print("AUC:", roc_auc_score(diag_numeric, preds))
print("Coefficients:", model.coef_)
```

# Insights

* AUC tells the strength or discriminatory power in the binary classification model
  * In this case, AUC ≈ 0.99 means the logistic model is almost flawles in the separation between Malignant and Benign classes,
* The Logistic Regression model will look like this:

<br>

$$ \text{Risk of malignant tumor} = 2.03*PC_1 - 1.12* PC_2 $$

<br>
<br>

* The <font color="green"> model's coefficeint </font>means:
  * <font color="green"> +2.03: </font> as PC1 increases, the probability of malignancy <u>increases</u>
    * So tumors that are more structurally irregular are more likely to be malignant
  * <font color="green"> -1.12: </font> a higher value of PC2 <u>slightly reduces</u> the chances of malignancy
    * complexity does not alone drive malignancy


# <font color="green"> Storyline of PCs: </font>

> Together, these two Principal Components could give a clear message that most tumor sits somewhere in the space of _shape irregularity or distortion (PC1) and complexity (PC2)_.


> Other PCs (PC3 - PC10) are not very significant in determining malignancy, considering their individual variation is very low. For example, PC3 is less than 10% and continues to decrease until PC10 which is less than 2%
